```python
for i in range(1, 101):
    if i % 15 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
numbers = [x for x in range(100)]
squares  = [x*x for x in range(10)]            
evens    = [x for x in range(100) if x % 2 == 0]   

print(numbers) 
print(squares)
print( evens)
```
