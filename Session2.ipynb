```python
warehouse = {
    'laptop': {'stock': 25, 'price': 55000, 'supplier': 'TechCorp'},
    'mouse': {'stock': 100, 'price': 850, 'supplier': 'GadgetHub'},
    'keyboard': {'stock': 45, 'price': 2200, 'supplier': 'TypeMaster'},
    'monitor': {'stock': 12, 'price': 18000, 'supplier': 'VisionPro'}
}

print("Original warehouse inventory:", warehouse)
```

```python
#Update

warehouse['laptop']['stock'] = 20

stock_updates = {'stock': 120, 'location': 'Rack-A'}
warehouse['mouse'].update(stock_updates)

print(warehouse)
```

```python
#Delete/Clear/Remove
warehouse_temp = {
    'laptop': {'stock': 25, 'price': 55000},
    'mouse': {'stock': 100, 'price': 850},
    'keyboard': {'stock': 45, 'price': 2200}
}


removed_price = warehouse_temp['laptop'].pop('price')
print(f"pop with argument from laptop removed {removed_price}: {warehouse_temp['laptop']}")


last_item = warehouse_temp['mouse'].popitem()
print(f"popitem() removed {last_item}: {warehouse_temp['mouse']}")


del warehouse_temp['keyboard']['price']
print("After delete", warehouse_temp['keyboard'])


warehouse_temp['keyboard'].clear()
print("After clear():", warehouse_temp['keyboard'])
```

```python
#IfElse
arr=[10,45,36,45,68,98,74,15,245,632,54,15,2,23,521,152,543]
arr.sort()
print("Sorted array is:",arr)

searchitem=245
n=len(arr)
first=0
while first<=n:
    i=(first+n)//2
    if searchitem==arr[i]:
        print("Found at", i)
        break
    elif searchitem<arr[i]:
        n=i-1
    else:
        first=i+1
else:
    print("Not Found")
        
    
```

```python
warehouse = {'laptop':10, 'mouse':50, 'keyboard':20}

arr = [
    {'id':'ORD001', 'item':'laptop', 'qty':3, 'pay':'completed', 'pin':570012, 'value':165000},
    {'id':'ORD002', 'item':'mouse', 'qty':15, 'pay':'completed', 'pin':570023, 'value':12750}
]

for order in arr:
    item, qty, pay, pin, value = order['item'], order['qty'], order['pay'], order['pin'], order['value']
    
    if qty > warehouse[item]:
        print(f"Order {order['id']}: OUT OF STOCK")
    elif pay != 'completed':
        print(f"Order {order['id']}: PAYMENT PENDING")
    elif not (570001 <= pin <= 570030):
        print(f"Order {order['id']}: BAD AREA")
    elif value > 50000:
        print(f"Order {order['id']}: MANAGER NEEDED")
    else:
        warehouse[item] -= qty
        print(f"Order {order['id']}: FULFILLED ({warehouse[item]} left)")
```

```python
#Tuple

menu = (
    ('Filter Coffee', 25),
    ('Masala Dosa', 80),
    ('Idli Vada', 60),
    ('Pongal', 70)
)

for item, price in menu:
    print(f"  {item}: Rs.{price}")


item_names = [name for name, price in menu]     
total_items = len(menu)                     
avg_price = sum(price for name, price in menu) / total_items

print(f"\n{total_items} items, Avg: ₹{avg_price}")
```
